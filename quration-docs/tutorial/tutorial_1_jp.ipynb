# プリセット回路の生成

このドキュメントでは quration で用意しているプリセットの回路の生成方法について説明します。

以下の例では、 `quration-algorithm` をコンパイルして得られたアルゴリズムの生成プログラムを呼び出します。このため、`README.md`に従い、`./build/examples/` フォルダに`create_qpe`などの実行ファイルが生成されていることを確認してください。以下では、`./build/examples/`を一時的にパスに追加し呼び出します。

実行ファイルが生成されるディレクトリはOSに依存しており、以下のコードはLinuxを対象としています。
例えば、Windowsの場合は`./build/bin/`フォルダに実行ファイルが生成されるため、適宜切り替えてください。

また、以降のコマンド例において JSON ファイルを指定するコマンドでは、`quration-algorithm/examples/data` の下にあるデフォルトの JSON ファイルを使用するため、`quration-algorithm/examples/data` へのパスを取得します。

````python
import os
import pathlib
import platform

project_root = pathlib.Path("../..").resolve()
if platform.system() == "Windows":
    algorithm_generator_path = project_root / "build" / "bin"
elif platform.system() == "Darwin": 
    algorithm_generator_path = project_root / "build" / "examples"
elif platform.system() == "Linux":
    algorithm_generator_path = project_root / "build" / "examples"
else:
    raise ValueError("Unsupported platform")

sample_data_dir = pathlib.Path("../..").resolve() / "quration-algorithm" / "examples" / "data"

os.environ["PATH"] = str(algorithm_generator_path) + os.pathsep + os.environ.get("PATH", "")
````

## QPE

QPEの回路の生成には `build/examples/create_qpe` という実行ファイルを実行します。

  まず、生成に必要となる情報が書かれた以下のフォーマットの JSON ファイルを用意します。  

```json
{
    "paulis": [
        [["Z"], ["I"]],
        [["Z"], ["Z"]],
        [["Y", "X"], ["I"]],
        [["Y", "X"], ["X", "Y", "X", "Y", "X"]]
    ],                                              // Pauli 文字列の配列
    "lcu_coefficients": [0.25, 0.25, 0.25, 0.25],   // LCU 係数
    "system_size": 2,                               // システムのビット数
    "hadamard_size": 3,                             // 位相のビット精度
    "sub_bit_precision": 3                          // PREPARE 回路のビット精度
}
```

次のコマンドを実行すると、QPE の回路の中間表現の JSON ファイルが出力されます。  
ここでは、`quration_algorithm/examples/data/sample_qpe.json` を指定する場合のコマンド例を示しています。

````python
!create_qpe --file {sample_data_dir / "sample_qpe.json"} --out qpe.json
````

この実行ファイルのコマンドラインオプションは `--help` オプションで確認することができます。

- `--file`: 入力するJSONファイルのパス
  - 指定しない場合は `quration_algorithm/examples/data/sample_qpe.json` がデフォルトの入力として設定されます。  
- `--out`: 出力する JSON ファイルのパス
  - 指定しない場合は `out.json` という名前で出力されます。  
- `--inline`: モジュール内のすべての回路をインライン展開して出力するオプション

## Trotter expansion

トロッター展開を用いた Hamiltonian の時間発展をシミュレーションする回路の生成には `examples/create_trotter` という実行ファイルを実行します。 


入力で指定する Hamiltonian は Pauli 文字列から構成されることを仮定します。  
Hamiltonian は次のようなフォーマットの JSON ファイルを作成して量子ビット数と Hamiltonian を構成する Pauli 文字列と係数を指定します。  

```json
{
    "num_qubits": 3,
    "pauli_terms": [
        {
            "coeff": 1.0,
            "pauli_string": {
                "0": "X"
            }
        },
        {
            "coeff": 1.0,
            "pauli_string": {
                "0": "Z",
                "1": "Z"
            }
        },
        {
            "coeff": 1.0,
            "pauli_string": {
                "1": "X"
            }
        },
        {
            "coeff": 1.0,
            "pauli_string": {
                "1": "Z",
                "2": "Z"
            }
        },
        {
            "coeff": 1.0,
            "pauli_string": {
                "2": "X"
            }
        },
        {
            "coeff": 1.0,
            "pauli_string": {
                "2": "Z",
                "0": "Z"
            }
        }
    ]
}
```

次のようにコマンドを実行すると、指定した Hamiltonian の時間発展をシミュレートする回路の JSON ファイルが出力されます。  
ここでは、`quration_algorithm/examples/data/1d_ising_hamiltonian.json` を指定する場合のコマンド例を示しています。

````python
!create_trotter --file {sample_data_dir / "1d_ising_hamiltonian.json"} --time 10 --num_trotter_steps 20 --out trotter.json
````

この実行ファイルのコマンドラインオプションは `--help` オプションで確認することができます。

- `--file`: 入力とする Hamiltonian のJSONファイルのパス
  - 指定しない場合、`quration_algorithm/examples/data/1d_ising_hamiltonian.json` がデフォルトの入力として設定されます。
- `--time`: 時間発展する時間
  - 指定しない場合はデフォルト値として 1.0 が使用されます。
- `--num_trotter_steps`: トロッター展開のステップ数
  - 指定しない場合はデフォルト値として 1 が使用されます。  
- `--out`: 出力する JSON ファイルのパス
  - 指定しない場合は `out.json` という名前で出力されます。  
- `--inline`: モジュール内のすべての回路をインライン展開して出力するオプション
